# 16.7.5

```python
x, y, z = sp.symbols("x y z")
sp.solveset(x / y + y / z + z / x - 1, x)
```

```python
F = x / y + y / z + z / x - 1
F
```

```python
z_x = -sp.diff(F,x) / sp.diff(F,z)
sp.simplify(z_x)
```

```python
z_y = -sp.diff(F,y) / sp.diff(F,z)
sp.simplify(z_y)
```

```python
z_xx = sp.diff(z_x,x)
sp.simplify(z_xx)
```

```python
z_xy = sp.diff(z_x,y) + sp.diff(z_x, z) * z_y
z_xy = sp.simplify(z_xy)
```

```python
z_yx = sp.diff(z_y,x) + sp.diff(z_y, z) * z_x
z_yx = sp.simplify(z_yx)
```

```python
z_yx == z_xy
```

```python
z_yy = sp.diff(z_y,y)
sp.simplify(z_yy)
```

# 17.2.11

```python
f1_f, f2_f, f3_f = sp.Function("f1"), sp.Function("f2"), sp.Function("f3")
g1_f, g2_f, g3_f = sp.Function("g1"), sp.Function("g2"), sp.Function("g3")
t = sp.symbols("t")
f1, f2, f3 = f1_f(t), f2_f(t), f3_f(t)
g1, g2, g3 = g1_f(t), g2_f(t), g3_f(t)
```

```python
f = sp.Matrix([f1, f2, f3])
g = sp.Matrix([g1, g2, g3])
```

```python
LHS = f.cross(g).diff(t)
sp.simplify(LHS)
```

```python
RHS = f.diff(t).cross(g) + f.cross(g.diff(t))
sp.simplify(RHS)
```

```python
LHS == RHS
```
